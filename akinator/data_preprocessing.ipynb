```python
import pandas as pd
import numpy as np
from datetime import datetime

# Assuming df is your DataFrame and 'date_of_birth' is the column with dates in 'yyyy-mm-dd' format
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])

# Current date
current_date = datetime(2023, 12, 18)

# Calculate age in years, round to the nearest whole number, and convert to integer
df['date_of_birth'] = np.round((current_date - df['date_of_birth']).dt.days / 365.25).astype(int)

# Rename the column to 'Age'
df.rename(columns={'date_of_birth': 'Age'}, inplace=True)

# Display the modified DataFrame
print(df)
```

```python
conditions_male = [
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 125),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 135),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 145),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 155),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 170),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 185),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 205),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] <= 225),
    (df['Gender'] == 'Male') & (df['weight_in_lbs'] > 225)
]

conditions_female = [
    (df['Gender'] == 'Female') & (df['weight_in_lbs'] <= 115),
    (df['Gender'] == 'Female') & (df['weight_in_lbs'] <= 125),
    (df['Gender'] == 'Female') & (df['weight_in_lbs'] <= 135),
    (df['Gender'] == 'Female') & (df['weight_in_lbs'] <= 145)
]

# Define the corresponding weight classes
weight_classes_male = ["in Men's Strawweight division", "in Men's Flyweight division", "in Men's Bantamweight division", "in Men's Featherweight division",
                       "in Men's Lightweight division", "in Men's Welterweight division", "in Men's Middleweight division", "in Men's Light Heavyweight division", "in Men's Heavyweight division"]

weight_classes_female =  ["in Women's Strawweight division", "in Women's Flyweight division", "in Women's Bantamweight division", "in Women's Featherweight division"]

# Create the 'Weight Class' column based on conditions
df['Weight Class'] = np.select(conditions_male, np.array(weight_classes_male), default=np.select(conditions_female,np.array(weight_classes_female), default='Unknown'))

# Display the modified DataFrame
print(df)
```
